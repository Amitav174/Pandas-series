Creating Dataframe

```python
#using list
student_data=[
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [60,50,2]
]
pd.DataFrame(student_data,columns=['iq','marks','package'])
```

```python
#using dict
student_dict={
    'iq':[100,90,120,80],
    'marks':[80,70,50,40],
    'package':[10,7,14,2]
}
students=pd.DataFrame(student_dict)
students
```
