What is pandas?

Pandas is a fast ,pwerful ,felxibel and easy to use open source data analysis and manipulation tool built on top of the python programming language

Pandas Series

pandas series like a column in a table. it is a 1-D array holding data of any type

importing pandas

```python
import numpy as np #it is good to import numpy
import pandas as pd
```

series from list

```python
#string
country =['Bnagladesh','India','Pakistan','Sri-lanka','Nepal']
pd.Series(country)
```

```python
#int
runs=[13,24,56,78,100]
pd.Series(runs)
```

Custom index

```python
marks=[67,57,89,100]
subjects=['math','english','chem','science']
pd.Series(marks,index=subjects)
```

setting a name

```python
marks=[67,57,89,100]
subjects=['math','english','chem','science']
pd.Series(marks,index=subjects,name="Amitav's marks")
```

Series from dictionary

```python
marks={
    'maths':67,
    'english':57,
    'science':89,
    'bangla':80
}
marks_series=pd.Series(marks,name="Amitav's marks")
marks_series
```
